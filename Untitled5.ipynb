```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
# import tushare as ts
from sklearn.preprocessing import StandardScaler
from torch.utils.data import TensorDataset
from tqdm import tqdm
```

```python
seq_size=5
input_size=13
hidden_size=8
batch_size=16
num_layers=3

epochs = 10
best_loss = 0
save_path = './best_model.pth'
```

```python
df = pd.read_csv("./jena_climate_2009_2016.csv", index_col = 0)
```

```python
df.shape
```

```python
df2=df.drop('T (degC)',axis=1)
```

```python
scaler = StandardScaler()
scaler_model = StandardScaler()

data = scaler_model.fit_transform(np.array(df2))
data_y=scaler.fit_transform(np.array(df['T (degC)']).reshape(-1, 1))
```

```python
def split_data(data,data_y,timestep,input_size):
    dataX_train = []  # 保存X_Train
    dataY_train = []  # 保存Y_Train
    dataX_test=[]
    dataY_test=[]
    
    train_size = int(np.round(0.8 * data.shape[0]))   #训练集数量
    test_size=len(data)-train_size                    #测试集数量

    
    for index in range(train_size - timestep):
        dataX_train.append(data[index: index + timestep])      
        dataY_train.append(data_y[index :index+ timestep])     #()
    
    X_train=np.array(dataX_train)                      #(batch,seq_len,input_size)
    Y_train=np.array(dataY_train)   #.reshape(-1,1)        #(batch*seq_len,1)
    
    for index in range(train_size,len(data)-timestep,timestep):
        dataX_test.append(data[index:index+timestep])
        dataY_test.append(data_y[index:index+timestep])
        
    X_test=np.array(dataX_test)                      #(batch,seq_len,input_size)
    Y_test=np.array(dataY_test)  #.reshape(-1,1)        #(batch*seq_len,1)
    
    return [X_train,Y_train,X_test,Y_test]
```

```python
x_train, y_train, x_test, y_test = split_data(data,data_y,seq_size, input_size)
```

```python
x_train_tensor = torch.from_numpy(x_train).to(torch.float32)
y_train_tensor = torch.from_numpy(y_train).to(torch.float32)
x_test_tensor = torch.from_numpy(x_test).to(torch.float32)
y_test_tensor = torch.from_numpy(y_test).to(torch.float32)
```

```python
train_data = TensorDataset(x_train_tensor, y_train_tensor)
test_data = TensorDataset(x_test_tensor, y_test_tensor)
```

```python
train_loader = torch.utils.data.DataLoader(train_data,
                                           batch_size,
                                           True)

test_loader = torch.utils.data.DataLoader(test_data,
                                          batch_size,
                                          False)
```

```python
class Model(nn.Module):
    def __init__(self,input_size,hidden_size,num_layers):
        super().__init__()
        
        self.input_size=input_size
        self.hidden_size=hidden_size
        self.num_layers=num_layers
        
        self.rnn=torch.nn.RNN(input_size=self.input_size
                              ,hidden_size=self.hidden_size
                              ,num_layers=self.num_layers
                              ,batch_first=True)
        
        self.fc=torch.nn.Linear(hidden_size,1)
        
    def forward(self,x):
        hidden=torch.zeros(self.num_layers,x.shape[0],self.hidden_size)
        x,_=self.rnn(x,hidden)
        x=self.fc(x)
        return x.view(-1,1)
```

```python
net=Model(input_size,hidden_size,num_layers)
```

```python
loss_function = nn.MSELoss()  # 定义损失函数
optimizer = torch.optim.Adam(net.parameters(), lr=0.015)  # 定义优化器
```

```python
for epoch in range(10):
    net.train()
    running_loss=0
    train_bar = tqdm(train_loader)  # 形成进度条
    for data in train_bar:
#         print("*"*20)
        x_train, y_train = data  # 解包迭代器中的X和Y
        optimizer.zero_grad()
        y_pred=net(x_train)
        loss=loss_function(y_pred,y_train.view(-1,1))
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        train_bar.desc = "train epoch[{}/{}] loss:{:.6f}".format(epoch + 1,
                                                                 epochs,
                                                                 loss)
    net.eval()
    test_loss = 0
    with torch.no_grad():
        test_bar = tqdm(test_loader)
        for data in test_bar:
            x_test, y_test = data
            y_test_pred = net(x_test)
            test_loss = loss_function(y_test_pred,y_test.view(-1,1))
        print("1"*20)
            
    if test_loss < best_loss:
        best_loss = test_loss
        torch.save(model.state_dict(), save_path)

print('Finished Training')
```
